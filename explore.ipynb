## 1. Clean your Datasets

For the first part of your project, you will apply the same data cleaning steps to your "data/control.txt" and "data/treatment.txt" files. By the end of this step, you should have two lists of floats that match the numbers you have in each respective file.

You will implement your `clean_data` function in your `cleaner.py` module and import the module in the code-block below.

```python
# TODO: import all the packages that you will need for this notebook here

import statistics as stats
import matplotlib.pyplot as plt

import cleaner
```

```python
# TODO: read your text files into your code using file I/O
control_file = open("data/control.txt")
treatment_file = open("data/treatment.txt")

control_data = control_file.readlines()
treatment_data = treatment_file.readlines()


# TODO: clean both list objects that you created above using the 'clean_data' function

clean_control = cleaner.clean_data(control_data)
clean_treatment = cleaner.clean_data(treatment_data)

print("Clean Control Data: ", clean_control)
print("Clean Treatment Data: ", clean_treatment)
```

## 2. Explore your Datasets

Now that you have your list objects created and cleaned, let's explore the descriptive statistics of these two files and make subsequent observations on our data.

```python
# TODO: Calculate the standard deviation of both the control and treatment group
stdev_control = stats.stdev(clean_control)
stdev_treatment = stats.stdev(clean_treatment)

# TODO: print your findings
print("Standard Deviation of Control Data: ", stdev_control)
print("Standard Deviation of Treatment Data: ", stdev_treatment)
```

```python
# TODO: Calculate the mean of both the control and treatment group
mean_control = stats.mean(clean_control)
mean_treatment = stats.mean(clean_treatment)

# TODO: print your findings
print("Mean of Control: ", mean_control)
print("Mean of Treatment: ", mean_treatment)
```

```python
# TODO: Are there any outliers in your control group? (use 2 * stdev to find these outliers)
outliers_control = []
control_upper_limit = mean_control + (stdev_control * 2)
control_lower_limit = mean_control - (stdev_control * 2)

# sanity check:
print("Lower Boundary: ", round(control_lower_limit, 2))
print("Control Mean: ", round(mean_control, 2))
print("Upper Boundary: ", round(control_upper_limit, 2))

# Create the list of control outliers:
for value in clean_control:
    if value > control_upper_limit or value < control_lower_limit:
        outliers_control.append(value)
    

print("Control Outliers :", outliers_control)
```

```python
# TODO: Are there any outliers in your treatment group?
outliers_treatment = []
treatment_upper_limit = mean_treatment + (stdev_treatment * 2)
treatment_lower_limit = mean_treatment - (stdev_treatment * 2)

#sanity check:
print("Lower Boundary: ", round(treatment_lower_limit, 2))
print("Treatment Mean: ", round(mean_treatment, 2))
print("Upper Boundary: ", round(treatment_upper_limit, 2))

# Create list of outliers for the treatment data:
for value in clean_treatment:
    if value > treatment_upper_limit or value < treatment_lower_limit:
        outliers_treatment.append(value)

print("Treatment Outliers :", outliers_treatment)
```

## 3. Create Data Visualizations

For this final step, generate data visualizations to view the dispersion of both of your groups.

Make sure that the two visualizations you create are on the same figure. 

```python
# TODO: create a data visualization to view differences in dispersion and mean
plt.hist(clean_control, alpha = 0.85, label='Control')
plt.hist(clean_treatment, alpha = 0.85, label='Treatment')
plt.xlabel('Minutes')
plt.ylabel('Number of Employees')
plt.title('Viewing Times of Control Group vs Treatment Group')
plt.legend(loc='best')
```
